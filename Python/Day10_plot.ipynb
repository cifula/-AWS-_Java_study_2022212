```python
import matplotlib.pyplot as plt
import numpy as np

fig, axe = plt.subplots(1)
# axe.plot(np.arange(10))

x = np.arange(10)
y = x * 5 + 10
axe.plot(x,y)
```

```python
# x = np.arange(10)
x = np.linspace(2, 4, 10)
y = 2 * x ** 2 + 5

_, axe = plt.subplots()
axe.plot(x,y)
```

```python
import math
np.pi
np.exp(5)
mu = 0
sig = 1
x = np.linspace(-5, 5, 100)

y = 1 / sig * math.sqrt(2 * np.pi) * np.exp(-((x - mu) ** 2) / 2 * sig **2 )

_, axe = plt.subplots()
axe.plot(x,y)
```

```python
from scipy import special
y = 1 / 2 * (1 + special.erf((x - mu) / sig * math.sqrt(2)))

_, axe = plt.subplots()
axe.plot(x,y)
```
