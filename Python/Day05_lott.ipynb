```python
csv_data = []
with open('data/lott.csv') as f:
    for line in f:
        csv_data.append(line[:-1].split(','))
print(csv_data[:3])
int(csv_data[0][-1])
```

```python
# 1. 전체 회차의 번호별 빈도수 (mode)

# flat
lott_num = []

for num in csv_data:
    for i in range(7):
        lott_num.append(num[i])
        
lott_set = set(lott_num)

lott_freq = []


for num in lott_set:
    key = int(num)
    value = lott_num.count(num)
    lott_freq.append((key, value))
    
# 정렬
lott_freq.sort(key = lambda x : x[0])
```

```python
# 2. 회차별 - 평균값, 표준편차
import math

def mean(num_list):
    sum_value = 0
    for s in num_list:
        sum_value += s
    return sum_value / len(num_list)

def dev(num_list):
    m = mean(num_list)
    return [x - m for x in num_list]

def var(num_list):
    n = len(num_list)
    d = dev(num_list)
    return sum([x*x for x in d])/(n-1)

def stdev(num_list):
    return math.sqrt(var(num_list))

lott_summary = []

for i, num in enumerate(csv_data):
    i += 1
    int_num = []
    for j in range(7):
        int_num.append(int(num[j]))
    lott_summary.append((i, mean(int_num), stdev(int_num)))
```

```python
# 1번 출력
for freq in lott_freq:
    print(freq[0], ":", freq[1])
# print(lott_mode)
```

```python
# 2번 출력
for summary in lott_summary:
    print(summary[0], "회차", "평균:", "{:.2f}".format(summary[1]), "표준편차:", "{:.2f}".format(summary[2]))
```
